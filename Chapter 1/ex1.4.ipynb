```python
from thinkdsp import read_wave
wave = read_wave('186942__lemoncreme__piano-melody.wav')
wave.normalize()
wave.make_audio()
```

```python
wave.plot()
```

```python
def stretch(wave, factor):
    wave.ts *= factor
    wave.framerate /= factor #higher pitch, faster
```

```python
stretch(wave, 0.3)
wave.make_audio()
```

```python
wave.plot()
```
