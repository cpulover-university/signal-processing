440Hz triangle wave with duration 0.01 seconds:

```python
triangle = TriangleSignal().make_wave(duration=0.01)
triangle.plot()
decorate(xlabel='Time (s)')
decorate(ylabel='Amplitude')
```

```python
spectrum = triangle.make_spectrum()
spectrum.hs[0]
```

The first element of the spectrum is a complex number close to zero.

```python
spectrum.hs[0] = 100
triangle.plot(color='red')
spectrum.make_wave().plot()
decorate(xlabel='Time (s)')
decorate(ylabel='Amplitude')
```

The operation has the effect of adding a vertical offset to the wave.
