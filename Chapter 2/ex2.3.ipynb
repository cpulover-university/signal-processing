```python
square = SquareSignal(1500).make_wave(duration=1, framerate=10000)
square.make_spectrum().plot()
decorate(xlabel='Frequency (Hz)')
```

- The fundamental frequency is 1500 Hz 
- The 1st harmonic frequency is 1500*3=4500 Hz
- The 2nd harmonic frequency is 1500*5=7500 Hz, is aliased to 2500 Hz.
- The 3rd harmonic frequency is 1500*7=10500 Hz, is aliased twice to 500 Hz
- The 4th harmonic frequency is 1500*9=13500 Hz, is aliased twice to 3500 Hz.
- The 5th harmonic frequency is 1500*11=16500 Hz, ends up at 1500 Hz, so it contributes to the fundamental.

The remaining harmonics overlap with the above frequencies

When you listen to the wave, the fundamental pitch you perceive is the alias at 500 Hz.

Audio of this wave:

```python
square.make_audio()
```
