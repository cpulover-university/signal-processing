```python
signal = SawtoothSignal(freq=500)
wave = signal.make_wave(duration=1, framerate=20000)
```

Here's what the spectrum looks like.  The harmonics drop off like $1/f$.

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```

Apply the function in exercise 2.5

```python
def modify_spectrum(spectrum):
    spectrum.hs[0] = 0  # avoid division by 0
    spectrum.hs[1:] /= spectrum.fs[1:]
```

```python
spectrum.plot(color='red')
modify_spectrum(spectrum)
spectrum.scale(500)
spectrum.plot()
```

The harmonics now drop off like $1/f^2$.
