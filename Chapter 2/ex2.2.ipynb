```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np

class SawtoothSignal(Sinusoid):    
    def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles)
        ys = normalize(unbias(frac), self.amp)
        return ys
```

Compute the spectrum of a sawtooth wave:

```python
sawtooth = SawtoothSignal().make_wave(duration=1, framerate=40000)
sawtooth.make_spectrum().plot()
decorate(xlabel='Frequency (Hz)')
```

```python
sawtooth.make_spectrum().plot()

square = SquareSignal(amp=0.5).make_wave(duration=1, framerate=40000) 
square.make_spectrum().plot(color='red')

triangle = TriangleSignal(amp=0.8).make_wave(duration=1, framerate=40000)
triangle.make_spectrum().plot(color='yellow')

decorate(xlabel='Frequency (Hz)')
```

Unlike triangle ware and square wave, the sawtooth ware contains both even and odd harmonics.
In sawtooth, the harmonics amplitudes drop off as slowly as in square ware and more slowly than in triagle ware
