Function modifying spectrum:

```python
def modify_spectrum(spectrum):
    spectrum.hs[0] = 0  # avoid division by 0
    spectrum.hs[1:] /= spectrum.fs[1:]
```

A square wave sample:

```python
wave = SquareSignal(freq=350).make_wave(duration=1)
wave.make_audio()
```

Red is the original spectrum. Blue is the modified spectrum:

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=5000, color='red')
modify_spectrum(spectrum)
spectrum.scale(350)
spectrum.plot(high=5000)
decorate(xlabel='Frequency (Hz)')
```

The function reduces high frequencies, so it acts like a non-strict low pass filter, while low pass filter removes completely high frequencies.
Here's what it sounds like:

```python
filtered = spectrum.make_wave()
filtered.make_audio()
```
