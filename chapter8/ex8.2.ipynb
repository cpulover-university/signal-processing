## Exercise 8.2

In this chapter I claimed that the Fourier transform of a Gaussian
curve is also a Gaussian curve.  For discrete Fourier transforms,
this relationship is approximately true.

Try it out for a few examples.  What happens to the Fourier transform
as you vary `std`?

```python
import os, numpy as np, matplotlib.pyplot as plt, scipy.signal
from thinkdsp import decorate
if not os.path.exists('thinkdsp.py'):
    !wget https://github.com/cpulover/signal-processing/blob/master/thinkdsp.py
```

We define a function to plot the Gaussian window and its FFT side-by-side, given M and std value:

```python
def plot_gaussian_n_ftt(M, std):
    gaussian = scipy.signal.gaussian(M=M, std=std)
    gaussian /= sum(gaussian) #normalize the curve

    plt.subplot(1, 2, 1)
    plt.plot(gaussian)
    decorate(xlabel='Time')

    fft_gaussian = np.fft.fft(gaussian)
    fft_rolled = np.roll(fft_gaussian, M//2) #roll the frequencies around
    
    plt.subplot(1, 2, 2)
    plt.plot(np.abs(fft_rolled))
    decorate(xlabel='Frequency')
    plt.show()
```

```python
plot_gaussian_n_ftt(50,3)
```

The sample shows that Fourier transform of a Gaussian curve is also a Gaussian curve. Now we try changing `std` value a few times to see what will happen:

```python
plot_gaussian_n_ftt(50,5)
```

```python
plot_gaussian_n_ftt(50,10)
```

It can be noticed that when `std` value increases, the Gaussian becomes wider and its FFT becomes narrower.
